```python
#Substituir o dataframe atual, pelo novo apenas com as colunas que queremos manter
dados = dados[colunas_manter].reset_index(drop=True)
dados
```

Substituir os nomes das colunas utilizando o método **Replace**

Ao avaliar o dicionário de dados, percebe-se que muitas variáveis apresentam apenas SIM e o NÃO. para essas criaremos um mini dicionário.

```python
dados['MCLI_LOCAL'] = dados['MCLI_LOCAL'].replace({
    1:'Sim',
    2:'Não',
    9:'Ignorado'
})

dados
```

Buscar agora todas as séries/colunas que possuem a codificação SIM/NÃO, e vamos aplicar um laço FOR.

```python
colunas_sim_nao = ['MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA', 'CLI_EQUIMO', 'CLI_NECROS', 'CLI_LOCAL_',
                   'MCLI_SIST', 'CLI_NEURO', 'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL',
                   'CLI_OUTR_2', 'CON_SOROTE', 'COM_LOC', 'COM_SECUND', 'COM_NECROS', 'COM_COMPOR',
                   'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC',
                   'COM_CHOQUE', 'DOENCA_TRA']

for serie in colunas_sim_nao:
  dados[serie] = dados[serie].replace({1:'Sim', 2:'Não',9:'Ignorado'})

dados
```

Ajustar os formatos de Data

```python
colunas_data = ['DT_NOTIFIC','DT_SIN_PRI', 'DT_INVEST', 'ANT_DT_ACI', 'DT_OBITO', 'DT_ENCERRA', 'DT_DIGITA' ]

for serie in colunas_data:
  dados[serie] = pd.to_datetime(dados[serie], format='%Y%m%d', errors='coerce')

dados
```

### Reescrita das codificações do dicionário de dados.
Nesta etapa, deve-se reescrever TODAS as codificações do dicionário de dados. Para isso, usamos novamente a função REPLACE

```python
dados['ANT_TEMPO_'] = dados['ANT_TEMPO_'].replace({
    1: '0 - 1h',
    2: '1 - 3h',
    3: '3 - 6h',
    4: '6 - 12h',
    5: '12 e 24h',
    6: '24 e +h',
    9: 'Ignorado'
})

dados['ANT_LOCA_1'] = dados['ANT_LOCA_1'].replace({
    1: 'Cabeça',
    2: 'Braço',
    3: 'Ante-Braço',
    4: 'Mão',
    5: 'Dedo da Mão',
    6: 'tronco',
    7: 'Coxa',
    8: 'Perna',
    9: 'Pé',
    10: 'Dedo do Pé',
    99: 'Ignorado'
})

dados['CLI_TEMPO_'] = dados['CLI_TEMPO_'].replace({
    1 : 'Normal',
    2 : 'Alterado',
    9 : 'Não realizado'
})

dados['TP_ACIDENT'] = dados['TP_ACIDENT'].replace({
    1 : 'Serpente',
    2 : 'Aranha',
    3 : 'Escorpião',
    4 : 'Lagarta',
    5 : 'Abelha',
    6 : 'Outros',
    9 : 'Ignorado'
})

dados['ANI_SERPEN'] = dados['ANI_SERPEN'].replace({
      1: 'Botrópico',
      2: 'Crotálico',
      3: 'Elapídico',
      4: 'Laquético',
      5: 'Serpente não peçonhenta',
      9: 'Ignorado'
      })

dados['ANI_ARANHA'] = dados['ANI_ARANHA'].replace({
      1: 'Foneutrismo',
      2: 'Loxoscelismo',
      3: 'Latrodectismo',
      4: 'Outra aranha',
      9: 'Ignorado'
      })

dados['ANI_LAGART'] = dados['ANI_LAGART'].replace({
      1: 'Lonomia',
      2: 'Outra lagarta',
      9: 'Ignorado',
      })

dados['TRA_CLASSI'] = dados['TRA_CLASSI'].replace({
      1: 'Leve',
      2: 'Moderado',
      3: 'Grave',
      9: 'Ignorado'
      })

dados['EVOLUCAO'] = dados['EVOLUCAO'].replace({
      1: 'Cura',
      2: 'Óbito por acidente por animais peçonhentos',
      3: 'Óbito por outras causas',
      9: 'Ignorado'
      })

dados['CS_GESTANT'] = dados['CS_GESTANT'].replace({
      1: '1º Trimestre',
      2: '2º Trimestre',
      3: '3º Trimestre',
      4: 'Idade gestacional Ignorada',
      5: 'Não',
      6: 'Não se aplica',
      9: 'Ignorado'
      })
dados['CS_RACA'] = dados['CS_RACA'].replace({
      1: 'Branca',
      2: 'Preta',
      3: 'Amarela',
      4: 'Parda',
      5: 'Indígena',
      9: 'Ignorado',
      })
dados['CS_ESCOL_N'] = dados['CS_ESCOL_N'].replace({
      0: 'Analfabeto',
      1: '1ª a 4ª série incompleta do EF (antigo primário ou 1º grau)',
      2: '4ª série completa do EF (antigo primário ou 1º grau)',
      3: '5ª à 8ª série incompleta do EF (antigo ginásio ou 1º grau)',
      4: 'Ensino fundamental completo (antigo ginásio ou 1º grau)',
      5: 'Ensino médio incompleto (antigo colegial ou 2º grau)',
      6: 'Ensino médio completo (antigo colegial ou 2º grau)',
      7: 'Educação superior incompleta',
      8: 'Educação superior completa',
      9: 'Ignorado',
      10: 'Não se aplica'
      })



dados
```

## TRATAR A COLUNA NU_IDADE_N


A idade no SINAN é utilizada de forma codificada, obedecendo o seguinte critério:
1. Hora
2. Dia
3. Mês
4. Ano

Exemplos:

* 3007 - Sete meses
* 4035 - Trinta e cinco anos



```python
idades = []

for idade in dados['NU_IDADE_N']:
  if idade > 4000:
    idades.append(idade - 4000)
  else:
    idades.append(0)

dados['NU_IDADE_N_'] = idades

dados
```

# Decodificando ID_MUNICIP

A base de dados do SINAN usa o código do IBGE para nome dos municípios

Deve ser usada uma tabela correspondente de código de UF

```python
municipios = pd.read_csv('/content/municipios.csv')

municipios
```

Converter em dicionário no qual codigo do municipio : "Nome do Município"

```python
municipios['IBGE'] = (municipios['COD'] // 10).astype('int64')
municipios
```

```python
municipios = municipios[['IBGE', 'NOME']]
municipios = municipios.set_index('IBGE')

municipios
```

```python
#Transformar em dicionário
dicionario_municipios = municipios.to_dict()['NOME']
dicionario_municipios
```

```python
dados.info()
```

```python
#Usando a função map
dados['ID_MUNICIP_NOME'] = dados['ID_MUNICIP'].map(dicionario_municipios)
dados['ID_MN_RESI_NOME'] = dados['ID_MN_RESI'].map(dicionario_municipios)
dados['ANT_MUNIC_NOME'] = dados['ANT_MUNIC_'].map(dicionario_municipios)

dados
```

## Ocupação

```python
#Identifiquei que a coluna ID_OCUPA_N é do tipo object, então, preciso transforma-la em inteiro
## Para tal, vou criar uma nova série ID_OCUPA_N_INT
#pd.to_numeric(df['COD'], errors='coerce').astype('Int64')
dados['ID_OCUPA_N_INT'] = pd.to_numeric(dados['ID_OCUPA_N'], errors = 'coerce').astype('Int64')
dados['ID_OCUPA_N_INT'] = dados['ID_OCUPA_N_INT'].astype(str)
```

```python
ocupacao = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/csv/refs/heads/master/CBO2002%20-%20Ocupacao.csv')
ocupacao = ocupacao.set_index('CODIGO')
dicionario_ocupacao = ocupacao.to_dict()['TITULO']
dicionario_ocupacao

dados['ID_OCUPA_N_NOME'] = dados['ID_OCUPA_N_INT'].map(dicionario_ocupacao)
dados
```

```python
uf = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/csv/refs/heads/master/cod_uf.csv')
uf = uf.set_index('Código UF')
dicionario_UF = uf.to_dict()['UF']


dados['SG_UF_NOME'] = dados['SG_UF'].map(dicionario_UF)
dados['SG_UF_NOT_NOME'] = dados['SG_UF_NOT'].map(dicionario_UF)
dados['ANT_UF_NOME'] = dados['ANT_UF'].map(dicionario_UF)

dados
```
