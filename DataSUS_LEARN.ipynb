Escolher quais colunas manter, copianda a lista gerada na célula anterior, e apagand o nome das que não nos serão úteis.

```python
#Neste caso apenas removi 'TP_NOT', 'ID_AGRAVO'. este passo pode variar entre cada caso

colunas_manter = ['DT_NOTIFIC', 'SEM_NOT', 'NU_ANO', 'SG_UF_NOT',
       'ID_MUNICIP', 'ID_REGIONA', 'DT_SIN_PRI', 'SEM_PRI', 'ANO_NASC',
       'NU_IDADE_N', 'CS_SEXO', 'CS_GESTANT', 'CS_RACA', 'CS_ESCOL_N', 'SG_UF',
       'ID_MN_RESI', 'ID_RG_RESI', 'ID_PAIS', 'DT_INVEST', 'ID_OCUPA_N',
       'ANT_DT_ACI', 'ANT_UF', 'ANT_MUNIC_', 'ANT_LOCALI', 'ANT_TEMPO_',
       'ANT_LOCA_1', 'MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA', 'CLI_EQUIMO',
       'CLI_NECROS', 'CLI_LOCAL_', 'CLI_LOCA_1', 'MCLI_SIST', 'CLI_NEURO',
       'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL', 'CLI_OUTR_2',
       'CLI_OUTR_3', 'CLI_TEMPO_', 'TP_ACIDENT', 'ANI_TIPO_1', 'ANI_SERPEN',
       'ANI_ARANHA', 'ANI_LAGART', 'TRA_CLASSI', 'CON_SOROTE', 'NU_AMPOLAS',
       'NU_AMPOL_1', 'NU_AMPOL_8', 'NU_AMPOL_6', 'NU_AMPOL_4', 'NU_AMPO_7',
       'NU_AMPO_5', 'NU_AMPOL_9', 'NU_AMPOL_3', 'COM_LOC', 'COM_SECUND',
       'COM_NECROS', 'COM_COMPOR', 'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM',
       'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC', 'COM_CHOQUE', 'DOENCA_TRA',
       'EVOLUCAO', 'DT_OBITO', 'DT_ENCERRA', 'DT_DIGITA']
```

```python
#Substituir o dataframe atual, pelo novo apenas com as colunas que queremos manter
dados = dados[colunas_manter].reset_index(drop=True)
dados
```

Substituir os nomes das colunas utilizando o método **Replace**

Ao avaliar o dicionário de dados, percebe-se que muitas variáveis apresentam apenas SIM e o NÃO. para essas criaremos um mini dicionário.

```python
dados['MCLI_LOCAL'] = dados['MCLI_LOCAL'].replace({
    1:'Sim',
    2:'Não',
    9:'Ignorado'
})

dados
```

Buscar agora todas as séries/colunas que possuem a codificação SIM/NÃO, e vamos aplicar um laço FOR.

```python
colunas_sim_nao = ['MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA', 'CLI_EQUIMO', 'CLI_NECROS', 'CLI_LOCAL_',
                   'MCLI_SIST', 'CLI_NEURO', 'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL',
                   'CLI_OUTR_2', 'CON_SOROTE', 'COM_LOC', 'COM_SECUND', 'COM_NECROS', 'COM_COMPOR',
                   'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC',
                   'COM_CHOQUE', 'DOENCA_TRA']

for serie in colunas_sim_nao:
  dados[serie] = dados[serie].replace({1:'Sim', 2:'Não',9:'Ignorado'})

dados
```

Ajustar os formatos de Data

```python
colunas_data = ['DT_NOTIFIC','DT_SIN_PRI', 'DT_INVEST', 'ANT_DT_ACI', 'DT_OBITO', 'DT_ENCERRA', 'DT_DIGITA' ]

for serie in colunas_data:
  dados[serie] = pd.to_datetime(dados[serie], format='%Y%m%d', errors='coerce')

dados
```
